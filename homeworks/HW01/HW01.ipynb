```python
import numpy as np

# 1. Создание массива
arr1 = np.array([1, 2, 3, 4])
print("arr1:", arr1)

# Индексирование
print("arr1[1] =", arr1[1])

# Простое вычисление
print("arr1 * 10 =", arr1 * 10)


# 2. Массив из диапазона
arr2 = np.arange(0, 10, 2)
print("\narr2:", arr2)

# Индексирование
print("arr2[-1] =", arr2[-1])


# 3. Двумерный массив
arr3 = np.array([[1, 2], [3, 4]])
print("\narr3:\n", arr3)

# Индексирование 2D
print("arr3[1,0] =", arr3[1, 0])

# Простое вычисление
print("arr3 + 5:\n", arr3 + 5)
```
