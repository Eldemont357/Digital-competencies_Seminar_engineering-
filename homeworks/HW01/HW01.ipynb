```python
import numpy as np

# -----------------------------
# 1. Создание массива (1D)
# -----------------------------
arr1 = np.array([1, 2, 3, 4])
print("arr1:", arr1)

# shape и dtype
print("arr1 shape:", arr1.shape)
print("arr1 dtype:", arr1.dtype)

# Индексирование
print("arr1[1] =", arr1[1])

# Простое вычисление
print("arr1 * 10 =", arr1 * 10)

# Статистики
print("arr1 mean =", arr1.mean())
print("arr1 sum  =", arr1.sum())


# -----------------------------
# 2. Массив из диапазона
# -----------------------------
arr2 = np.arange(0, 10, 2)
print("\narr2:", arr2)

# shape и dtype
print("arr2 shape:", arr2.shape)
print("arr2 dtype:", arr2.dtype)

# Индексирование
print("arr2[-1] =", arr2[-1])

# Статистики
print("arr2 mean =", arr2.mean())
print("arr2 sum  =", arr2.sum())


# -----------------------------
# 3. Двумерный массив
# -----------------------------
arr3 = np.array([[1, 2], [3, 4]])
print("\narr3:\n", arr3)

# shape и dtype
print("arr3 shape:", arr3.shape)
print("arr3 dtype:", arr3.dtype)

# Индексирование 2D
print("arr3[1,0] =", arr3[1, 0])

# Простое вычисление
print("arr3 + 5:\n", arr3 + 5)

# Статистики
print("arr3 mean =", arr3.mean())
print("arr3 sum  =", arr3.sum())
```
